```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.
```

#### Import data

```python
iris = datasets.load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

```python
y = iris.target
```

```python
sns.pairplot(data=X)
```

* We known targets have 3 identical value.

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(X)
```

```python
df = X.copy()
df['predict'] = kmeans.predict(X)
df['predict'] = df['predict'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
sns.pairplot(data=df, hue='predict')
```
