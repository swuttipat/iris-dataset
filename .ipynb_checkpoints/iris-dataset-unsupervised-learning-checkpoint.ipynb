### Transform dataset for Unsupervised Learning

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

#### K-Means

```python
from sklearn.cluster import KMeans
```

```python
iterates = []
inertias = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(X)
    iterates.append(i)
    inertias.append(kmeans.inertia_)

plt.plot(iterates, inertias)
plt.title('Optimal Number of Clusters')
plt.xlabel('Numbers of Cluster (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)

x = 'petal length (cm)'
y = 'petal width (cm)'

fig, axes = plt.subplots(nrows=1, ncols=5,figsize=(18,3), sharey=True)
for c, i in enumerate(range(2,6)):
    model = KMeans(i)
    y_hat = model.fit_predict(X)
    sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10',ax=axes[c])
    axes[c].legend().remove()
    axes[c].set_title(f"K = {i}")

# Targets
target = iris.target
c = 4
sns.scatterplot(data=X, x=x, y=y, hue=target, palette='tab10',ax=axes[c])
axes[c].legend().remove()
axes[c].set_title(f"Target")

plt.show()
```

#### Hierarchical clustering

```python
import numpy as np

from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)

x = 'petal length (cm)'
y = 'petal width (cm)'

fig, axes = plt.subplots(nrows=1, ncols=5,figsize=(18,3), sharey=True)
for c, i in enumerate(range(2,6)):
    model = AgglomerativeClustering(i)
    y_hat = model.fit_predict(X)
    sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10',ax=axes[c])
    axes[c].legend().remove()
    axes[c].set_title(f"K = {i}")

# Targets
target = iris.target
c = 4
sns.scatterplot(data=X, x=x, y=y, hue=target, palette='tab10',ax=axes[c])
axes[c].legend().remove()
axes[c].set_title(f"Target")

plt.show()
```
