```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

### Import Data

```python
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
df.head()
```

```python
sns.pairplot(data=df, hue='species', palette='tab10')
```

* petal length and petal width observation are clustering well, so we will make it as reference.

### Transform dataset for Unsupervised Learning

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

```python
def plot_clusters(model):
    X = pd.DataFrame(iris.data, columns=iris.feature_names)

    x = 'petal length (cm)'
    y = 'petal width (cm)'

    fig, axes = plt.subplots(nrows=1, ncols=5,figsize=(18,3), sharey=True)
    for c, i in enumerate(range(2,6)):
        model.set_params(n_clusters=i)
        y_hat = model.fit_predict(X)
        sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10',ax=axes[c])
        axes[c].legend().remove()
        axes[c].set_title(f"K = {i}")

    # Targets
    target = iris.target
    c = 4
    sns.scatterplot(data=X, x=x, y=y, hue=target, palette='tab10',ax=axes[c])
    axes[c].legend().remove()
    axes[c].set_title(f"Target")

    plt.show()
```

```python
def plot_pairplot(model,  params=None):
    
    if params==None:
        model.set_params()
    else:
        model.set_params(**params)
    
    X = pd.DataFrame(iris.data, columns=iris.feature_names)
    pp_df = df.copy()
    pp_df['predict'] = model.fit_predict(X)

    sns.pairplot(data=pp_df, hue='predict', palette='tab10')

    plt.show()
```

```python
 def plot_crosstab_heatmap(model, params=None):
    cross_hm_df = df.copy()
    if params==None:
        model.set_params()
    else:
        model.set_params(**params)
    cross_hm_df['predict'] = model.fit_predict(X)

    cross_hm_df = pd.crosstab(index=cross_hm_df['species'], columns=cross_hm_df['predict'])
    sns.heatmap(cross_hm_df, annot=True)
    
    plt.show()
    print(f"{model} accuracy : {cross_hm_df.max().sum()/150}")
    return cross_hm_df.max().sum()/150
```

### K-Means

```python
from sklearn.cluster import KMeans
```

**Elbow method (clustering)**
Using the "elbow" or "knee of a curve" as a cutoff point is a common heuristic in mathematical optimization to choose a point where diminishing returns are no longer worth the additional cost. In clustering, this means one should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data.

```python
iterates = []
inertias = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(X)
    iterates.append(i)
    inertias.append(kmeans.inertia_)

plt.plot(iterates, inertias)
plt.title('Optimal Number of Clusters')
plt.xlabel('Numbers of Cluster (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
km_params = {'n_clusters':3}
plot_clusters(KMeans())
km_score = plot_crosstab_heatmap(KMeans(), km_params)
```

### Spectral clustering

```python
from sklearn.cluster import SpectralClustering
```

```python
spc_params = {'n_clusters':3}
plot_clusters(SpectralClustering())
spc_score = plot_crosstab_heatmap(SpectralClustering(), spc_params)
```

### Hierarchical clustering

```python
import numpy as np

from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

```python
params = {'n_clusters':3}
model = AgglomerativeClustering()
print('Hierarchical Clustering')
plot_clusters(model)
hc_score = plot_crosstab_heatmap(model, params)
```

```python
pd.DataFrame({'Model':['Kmeans', 'Spectral', 'Hierarchical'], 'Accuracy':[km_score, spc_score, hc_score]})
```
