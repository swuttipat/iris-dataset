# Iris Data Set
This is perhaps the best known database to be found in the pattern recognition literature. Fisher's paper is a classic in the field and is referenced frequently to this day. (See Duda & Hart, for example.) The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant. One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

# Purpose
* Get familiar with unsupervised learning algorithm
* To test which model is the best fit for clustering given datasets

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
```

## Import Data

```python
iris = datasets.load_iris() # load data set
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
df.head()
```

Let use `sns.pairplot` to find insight of data between each features color by given species.

```python
sns.pairplot(data=df, hue='species', palette='tab10')
```

* The plot relation between **petal length** and **patal width** the observtion are cluster nice 3 groups.

## Transform dataset for Unsupervised Learning

For **unsupervised learning** the targets data are excluded, so we set `X` as features data that use for model trianing.

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
X.head()
```

## Build function for Models evaluate

```python
# Create scatter plot of clustering model start with n_cluster = 2 to 5 
# Compare the model result with original scatter plot

def plot_clusters(model):
    X = pd.DataFrame(iris.data, columns=iris.feature_names)

    x = 'petal length (cm)'
    y = 'petal width (cm)'
    
    # Plot scatter for n_cluster 2 to 5
    fig, axes = plt.subplots(nrows=1, ncols=5,figsize=(18,3), sharey=True)
    for c, i in enumerate(range(2,6)):
        model.set_params(n_clusters=i)
        y_hat = model.fit_predict(X)
        sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10',ax=axes[c])
        axes[c].legend().remove()
        axes[c].set_title(f"K = {i}")

    # Plot Targets
    target = iris.target
    c = 4
    sns.scatterplot(data=X, x=x, y=y, hue=target, palette='tab10',ax=axes[c])
    axes[c].legend().remove()
    axes[c].set_title(f"Target")

    plt.show()
```

```python
# Create heatmap plot show distribution between true values against predict values.
# Calulate model accuracy

def plot_crosstab_heatmap(model, params=None):
    
    # Setting parameter
    if params==None:
        model.set_params()
    else:
        model.set_params(**params)
        
    cross_hm_df = df.copy()
    cross_hm_df['predict'] = model.fit_predict(X) # train and predict model

    cross_hm_df = pd.crosstab(index=cross_hm_df['species'], columns=cross_hm_df['predict']) # create cross tab dataframe
    sns.heatmap(cross_hm_df, annot=True) # plot heatmap
    
    plt.show()
    print(f"{model} accuracy : {cross_hm_df.max().sum()/150}") # calculate accuracy
    return cross_hm_df.max().sum()/150
```

## K-Means

```python
from sklearn.cluster import KMeans
```

**Elbow method (clustering)**
Using the "elbow" or "knee of a curve" as a cutoff point is a common heuristic in mathematical optimization to choose a point where diminishing returns are no longer worth the additional cost. In clustering, this means one should choose a number of clusters so that adding another cluster doesn't give much better modeling of the data.<br>
In cluster analysis, the elbow method is a heuristic used in determining **the number of clusters in a data set.**

```python
# use Elbow method to find optimum number of clusters for this algorithm
iterates = []
inertias = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(X)
    iterates.append(i)
    inertias.append(kmeans.inertia_)

plt.plot(iterates, inertias)
plt.title('Optimal Number of Clusters')
plt.xlabel('Numbers of Cluster (k)')
plt.ylabel('Inertia')
plt.show()
```

```python
km_params = {'n_clusters':3}
plot_clusters(KMeans())
km_score = plot_crosstab_heatmap(KMeans(), km_params)
```

## Spectral clustering

```python
from sklearn.cluster import SpectralClustering
```

```python
spc_params = {'n_clusters':3}
plot_clusters(SpectralClustering())
spc_score = plot_crosstab_heatmap(SpectralClustering(), spc_params)
```

## Hierarchical clustering
**Hierarchical clustering** is a general family of clustering algorithms that build nested clusters by merging or splitting them successively. This hierarchy of clusters is represented as a tree (or dendrogram). The root of the tree is the unique cluster that gathers all the samples, the leaves being the clusters with only one sample. 

**The AgglomerativeClustering** object performs a hierarchical clustering using a bottom up approach: each observation starts in its own cluster, and clusters are successively merged together. The linkage criteria determines the metric used for the merge strategy

```python
import numpy as np

from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering
```

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=3)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

```python
params = {'n_clusters':3}
model = AgglomerativeClustering()
print('Hierarchical Clustering')
plot_clusters(model)
hc_score = plot_crosstab_heatmap(model, params)
```

#### DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
cross_db_df = df.copy()
dbscan = DBSCAN(eps=.5, min_samples=10)

cross_db_df['predict'] = dbscan.fit_predict(X)

cross_db_df = pd.crosstab(index=cross_db_df['species'], columns=cross_db_df['predict'])
sns.heatmap(cross_db_df, annot=True)
plt.show()
print(f"{model} accuracy : {cross_db_df.max().sum()/150}")

x = 'petal length (cm)'
y = 'petal width (cm)'


y_hat = dbscan.fit_predict(X)
sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10')
plt.show()
```

```python
db_params = {'eps':.5, 'min_samples':10}

dbscan = DBSCAN(**db_params)

x = 'petal length (cm)'
y = 'petal width (cm)'
y_hat = dbscan.fit_predict(X)

sns.scatterplot(data=X, x=x, y=y, hue=y_hat, palette='tab10')
plt.show()

db_score = plot_crosstab_heatmap(DBSCAN(),db_params)
```
